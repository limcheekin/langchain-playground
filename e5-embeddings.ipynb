# E5 Embeddings

Let's load the E5 Embedding class.

```python
from open.text.embeddings.huggingface import E5Embeddings
```

```python
encode_kwargs = {'normalize_embeddings': True}
embeddings = E5Embeddings(encode_kwargs = encode_kwargs)
```

```python
text = "This is a test document."
```

```python
query_result = embeddings.embed_query(text)
```

```python
len(query_result)
```

```python
doc_result = embeddings.embed_documents([text])
```

```python
embeddings.embed_documents(["This is line 1",
                            "A B C ...."])
```

```python
len(doc_result)
```

```python
len(doc_result[0])
```

```python
doc_result[0] == query_result
```
